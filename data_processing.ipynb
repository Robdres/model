```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier

def train_rf(X, y):
    X, y = X.astype(float), y.astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42)
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    print("Random Forest:\n", classification_report(y_test, model.predict(X_test)))
    return model, scaler

def predict_rf(model, scaler, new_patient):
    new_scaled = scaler.transform([new_patient])
    return model.predict(new_scaled)[0], model.predict_proba(new_scaled)[0][1]
```

```python
model, scaler = train_rf(database_clean, ckd)
new_patient = database_clean.iloc[np.random.randint(1,100)].values
pred, prob = predict_rf(model, scaler, new_patient)
print(f"Prediction: {'CKD' if pred else 'No CKD'}, Probability: {prob:.2f}")
```

```python
from xgboost import XGBClassifier

def train_xgb(X, y):
    X, y = X.astype(float), y.astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y)
    model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    model.fit(X_train, y_train)
    print("XGBoost:\n", classification_report(y_test, model.predict(X_test)))
    return model, scaler

def predict_xgb(model, scaler, new_patient):
    new_scaled = scaler.transform([new_patient])
    return model.predict(new_scaled)[0], model.predict_proba(new_scaled)[0][1]
```

```python
model, scaler = train_xgb(database_clean, ckd)
new_patient = database_clean.iloc[np.random.randint(1,100)].values
pred, prob = predict_xgb(model, scaler, new_patient)
print(f"Prediction: {'CKD' if pred else 'No CKD'}, Probability: {prob:.2f}")
```

```python
from lightgbm import LGBMClassifier

def train_lgbm(X, y):
    X, y = X.astype(float), y.astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y)
    model = LGBMClassifier(random_state=42)
    model.fit(X_train, y_train)
    print("LightGBM:\n", classification_report(y_test, model.predict(X_test)))
    return model, scaler

def predict_lgbm(model, scaler, new_patient):
    new_scaled = scaler.transform([new_patient])
    return model.predict(new_scaled)[0], model.predict_proba(new_scaled)[0][1]
```

```python
model, scaler = train_lgbm(database_clean, ckd)
new_patient = database_clean.iloc[np.random.randint(1,100)].values
pred, prob = predict_lgbm(model, scaler, new_patient)
print(f"Prediction: {'CKD' if pred else 'No CKD'}, Probability: {prob:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression

def train_logistic(X, y):
    X, y = X.astype(float), y.astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42)
    model = LogisticRegression()
    model.fit(X_train, y_train)
    print("Logistic Regression:\n", classification_report(y_test, model.predict(X_test)))
    return model, scaler

def predict_logistic(model, scaler, new_patient):
    new_scaled = scaler.transform([new_patient])
    return model.predict(new_scaled)[0], model.predict_proba(new_scaled)[0][1]
```

```python
model, scaler = train_logistic(database_clean, ckd)
new_patient = database_clean.iloc[np.random.randint(1,100)].values
pred, prob = predict_logistic(model, scaler, new_patient)
print(f"Prediction: {'CKD' if pred else 'No CKD'}, Probability: {prob:.2f}")
```

```python
from sklearn.svm import SVC

def train_svm(X, y):
    X, y = X.astype(float), y.astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42)
    model = SVC(kernel='rbf', probability=True)
    model.fit(X_train, y_train)
    print("SVM:\n", classification_report(y_test, model.predict(X_test)))
    return model, scaler

def predict_svm(model, scaler, new_patient):
    new_scaled = scaler.transform([new_patient])
    return model.predict(new_scaled)[0], model.predict_proba(new_scaled)[0][1]
```

```python
model, scaler = train_svm(database_clean, ckd)
new_patient = database_clean.iloc[np.random.randint(1,100)].values
pred, prob = predict_svm(model, scaler, new_patient)
print(f"Prediction: {'CKD' if pred else 'No CKD'}, Probability: {prob:.2f}")
```

```python
from sklearn.neural_network import MLPClassifier

def train_mlp(X, y):
    X, y = X.astype(float), y.astype(int)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42)
    model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
    model.fit(X_train, y_train)
    print("Neural Network (MLP):\n", classification_report(y_test, model.predict(X_test)))
    return model, scaler

def predict_mlp(model, scaler, new_patient):
    new_scaled = scaler.transform([new_patient])
    return model.predict(new_scaled)[0], model.predict_proba(new_scaled)[0][1]
```

```python
model, scaler = train_mlp(database_clean, ckd)
new_patient = database_clean.iloc[np.random.randint(1,100)].values
pred, prob = predict_mlp(model, scaler, new_patient)
print(f"Prediction: {'CKD' if pred else 'No CKD'}, Probability: {prob:.2f}")
```
