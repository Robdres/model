```python
from diagrams import Cluster, Diagram
from diagrams.gcp.analytics import BigQuery, Dataflow, PubSub
from diagrams.gcp.compute import AppEngine, Functions
from diagrams.gcp.database import BigTable
from diagrams.gcp.iot import IotCore
from diagrams.gcp.ml import AdvancedSolutionsLab
from diagrams.programming.framework import React,Flask,Vercel
from diagrams.programming.flowchart import Database
from diagrams.gcp.storage import GCS

with Diagram("KidneyTool", show=False) as diag:
    pubsub = PubSub("Selección\nde variables")

    with Cluster("Recolección de data"):
        [IotCore("AS400 IESSS") ] >> PubSub("Filtrado\nLimpieza") >> pubsub
        
    with Cluster("Modelo ML") as model:
        with Cluster("Python"):
            flow = AdvancedSolutionsLab("XGBooost")

        with Cluster("Data Lake"):
            flow2 = [ GCS("PKL model")]
            
    with Cluster("Web") as web:
        with Cluster("Python"):
            web2 = Flask("Flask")
        with Cluster("React"):
            web3 =  React("React")
        with Cluster("Servidor"):
            web4 =  [ Vercel("Vercel"),
                      Database("Firebase")]
                          
    pubsub >> flow
    flow>>flow2
    flow2>>web2
    web2>>web3
    web3>>web4
diag
```

```python
from diagrams import Diagram, Cluster
from diagrams.onprem.client import Users
from diagrams.programming.language import Javascript
from diagrams.onprem.compute import Server
from diagrams.generic.blank import Blank
from diagrams.aws.management import CloudformationTemplate
from diagrams.aws.management import OpsworksApps
from diagrams.aws.management import TrustedAdvisorChecklist

from diagrams.elastic.beats import Heartbeat
with Diagram("Arquitectura de la página web en React", direction="LR"):
    user = Users("Usuario")
    with Cluster("Aplicación Web (React)"):
        landing = OpsworksApps("Landing Page")
        form = CloudformationTemplate("Form")
        confirm = TrustedAdvisorChecklist("Confirmación")
        result = Heartbeat("Resultados")
        pdf = Heartbeat("Descarga PDF")
        user >> landing >> form >> confirm >> result >> pdf

    backend = Server("Backend / Modelo ML")
    result >> backend
    backend >> pdf
```
